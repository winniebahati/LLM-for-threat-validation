# **TP, FP, FN, TN- A, Boxplot**

```python
#Real and fake threats Group A

precision_list_A_GH = []
precision_list_A_k8 = []
precision_list_B_GH = []
precision_list_B_k8 = []
precision_list_C_GH = []
precision_list_C_k8 = []
precision_list_D_GH = []
precision_list_D_k8= []

recall_list_A_GH = []
recall_list_A_k8 = []
recall_list_B_GH = []
recall_list_B_k8 = []
recall_list_C_GH = []
recall_list_C_k8 = []
recall_list_D_GH = []
recall_list_D_k8 = []

TP_list_A_GH = []
FP_list_A_GH = []
FN_list_A_GH = []
TN_list_A_GH = []

for index, row in df6_A_GH.iterrows():

    real_threats = row[2:7]
    false_threats = row[7:12]

    # True Positive - Real threats being selected
    TP = 5 - real_threats.isnull().sum()
    TP_list_A_GH.append(TP)

    # False Negatives - Real threats not being selected
    FN = real_threats.isnull().sum()
    FN_list_A_GH.append(FN)

    # False Positive - Fake threats being selected
    FP = 5 - false_threats.isnull().sum()
    FP_list_A_GH.append(FP)

    # True Negative - Fake threats not being selected
    TN = 5 - FP
    TN_list_A_GH.append(TN)

    P = TP / (TP + FP)
    R = TP / (TP + FN)

    # print("Student_ID: " + str(int(row["Student_ID"])))
    # print("Precision: " + str(P))
    # print("Recall: " + str(R))

    precision_list_A_GH.append(P)
    recall_list_A_GH.append(R)

# precision_list_A = [ round(elem, 2) for elem in precision_list_A ]
# recall_list_A = [ round(elem, 2) for elem in recall_list_A ]
TP_list_A_GH  = [ round(elem, 2) for elem in TP_list_A_GH ]
FP_list_A_GH  = [ round(elem, 2) for elem in FP_list_A_GH ]
FN_list_A_GH  = [ round(elem, 2) for elem in FN_list_A_GH ]
TN_list_A_GH  = [ round(elem, 2) for elem in TN_list_A_GH ]
TP_TN_list_A_GH  =[round(elem, 2) for elem in TP_list_A_GH] + [ round(elem, 2) for elem in TN_list_A_GH  ]


TP_avg_A_GH = sum(TP_list_A_GH ) / len(TP_list_A_GH )
FP_avg_A_GH = sum(FP_list_A_GH ) / len(FP_list_A_GH )
FN_avg_A_GH = sum(FN_list_A_GH ) / len(FN_list_A_GH )
TN_avg_A_GH = sum(TN_list_A_GH ) / len(TN_list_A_GH )
TP_TN_Avg_A_GH = sum(TP_TN_list_A_GH ) / len(TP_TN_list_A_GH )

print(TP_TN_list_A_GH )
print(TP_avg_A_GH)
print(FP_avg_A_GH)
print(FN_avg_A_GH)
print(TN_avg_A_GH)
print(TP_TN_Avg_A_GH)
print(real_threats)
print(false_threats)
print(TP_list_A_GH)
print(TN_list_A_GH)
print(FP_list_A_GH)
print(FN_list_A_GH)
```
